```python
import os 
import cv2 as cv 
import numpy as np 

import pandas as pd 
import json 
import re 
import matplotlib.pyplot as plt 
```

```python
def reshape_poly_cordination(lst_1,lst_2):
    new_lst=[]
    for x,y in zip(lst_1,lst_2):
        new_lst.append((x,y))
    return new_lst
#---------------------------------------------------------
def reshape_json():
    all_json='annotation_data_1/ALL/annotations_all.json' 
    with open(all_json,'r') as fp :
        dict_json=json.load(fp)
    my_list=list(dict_json.keys())
    new_dict={}
    for num in range(len(my_list)):
        name=dict_json[my_list[num]]['regions'][0]['shape_attributes']['name']
        if name== 'polygon':
            
            all_points_x=dict_json[my_list[num]]['regions'][0]['shape_attributes']['all_points_x']

            all_points_y=dict_json[my_list[num]]['regions'][0]['shape_attributes']['all_points_y']
            final_list=reshape_poly_cordination(all_points_x,all_points_y)

            file_name=dict_json[my_list[num]]['filename']
    #         print(file_name)
            new_dict.update({file_name:[name,final_list]})
        if name=='ellipse':
            cx=int(dict_json[my_list[num]]['regions'][0]['shape_attributes']['cx'])
            cy=int(dict_json[my_list[num]]['regions'][0]['shape_attributes']['cy'])
            rx=int(dict_json[my_list[num]]['regions'][0]['shape_attributes']['rx'])
            ry=int(dict_json[my_list[num]]['regions'][0]['shape_attributes']['ry'])
            center_coordinations=(cx,cy)
            axes_length=(rx,ry)
            theta=dict_json[my_list[num]]['regions'][0]['shape_attributes']['theta']
            start_angle = 0
            end_angle = 360
            file_name=dict_json[my_list[num]]['filename']
            
            new_dict.update({file_name:[name,center_coordinations,axes_length,theta,start_angle,end_angle]})
        else : 
            continue
            
    return new_dict
a=reshape_json()
all_name=list(a.keys())

print(a['y114.jpg'])
```

```python
a=reshape_json()

def make_clean_data(img_name):
#     img_name='y101.jpg'
    
    masks_path='annotation_data_1/MASKS'
    all_image='annotation_data_1/ALL'
    images_path='annotation_data_1/IMAGES'
    data=a[img_name]
    if data[0]=='polygon' : 
        image=cv.imread(all_image+'/'+img_name,0)

        size=image.shape
        
        mask = np.zeros(size)
        end_points=np.array(data[1])
        cv.fillPoly(mask, pts=[end_points], color=(255))
        

    if data[0]=='ellipse':
        image=cv.imread(all_image+'/'+img_name,0)
        
        
        
        size=image.shape
        
        mask = np.zeros(size)     
        mask = cv.ellipse(mask, data[1], data[2],data[3], data[4], data[5], (255), -1)
        
    
    
    cv.imwrite(masks_path+'/'+img_name,mask)    
    cv.imwrite(images_path+'/'+img_name,image)
    return [mask,image]


    
```

```python
f=list(map(lambda x : make_clean_data(x),list(a.keys())))
print('it is done ')
```
